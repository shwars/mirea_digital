```python
import jturtle as t

def square(x):
  t.forward(x)
  t.right(90)
  t.forward(x)
  t.right(90)
  t.forward(x)
  t.right(90)
  t.forward(x)
  t.right(90)

def triangle(x):
  t.forward(x)
  t.right(120)
  t.forward(x)
  t.right(120)
  t.forward(x)
  t.right(120)

def house(x):
  square(x)
  t.forward(x)
  t.right(30)
  triangle(x)
  t.left(30)
  t.right(180)
  t.penup()
  t.forward(x//3)
  t.left(90)
  t.forward(x//3)
  t.pendown()
  square(x//3)

house(100)

t.done()
```

```python
def to_mins(h,m):
  return h*60+m

def to_hm(m):
  return m//60, m%60

def time_to_str(t):
  return f"{t[0]}:{t[1]:02}"

def str_to_time(s):
  h = s[:s.find(':')]
  m = s[s.find(':')+1:]
  return int(h),int(m)

def end_of_day(n,start="8:30",short_break=10,long_break=20,every=3):
  h,m = str_to_time(start)
  breaks = n-1
  adds = breaks//every
  return time_to_str(to_hm(to_mins(h,m)+n*45+
                          breaks*short_break+
                          adds*(long_break-short_break)))

end_of_day(4*2,short_break=0,every=2)
```

```python
def clock_hand_angle(s):
  f = float(s.replace(':','.'))
  d = f-int(f)
  d = d/6.0*10.0
  return (int(f)%12+d)/12*360

# clock_hand_angle("18:00")
```

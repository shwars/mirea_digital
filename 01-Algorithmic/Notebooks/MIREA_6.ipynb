```python
import sklearn 
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
def encode_gender(x):
  if x=='male':
    return 0
  else:
    return 1

df['gender'] = df['sex'].apply(encode_gender)
```

```python
df
```

```python
df.corr()
```

```python
features = ['age','pclass','gender','sibsp','parch']
df[features]
```

```python
import sklearn
import sklearn.model_selection 

X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(df[features],df['survived'],test_size=0.2)
```

```python
len(X_train), len(X_test)
```

```python
import sklearn.linear_model
model = sklearn.linear_model.LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
sklearn.metrics.ConfusionMatrixDisplay.from_estimator(model,X_test,Y_test)
```

```python
sklearn.metrics.accuracy_score(Y_test,Y_pred)
```

```python
list(zip(features,model.coef_[0]))
```

```python
import graphviz, sklearn.tree

model = sklearn.tree.DecisionTreeClassifier(max_depth=3).fit(X_train,Y_train)
# 
graphviz.Source(
    sklearn.tree.export_graphviz(model,feature_names=features,class_names=['no','yes'],
    filled=True, rounded=True, special_characters=True,impurity=False)
)
```

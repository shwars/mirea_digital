````python
import math

def solve(a,b,c):
  d = b*b-4*a*c
  if d>=0:
    x1 = (-b+math.sqrt(d))/(2*a)
    x2 = (-b-math.sqrt(d))/(2*a)
    return x1,x2
  else:
    return None,None

x1,x2 = solve(1,2,1)

if x1==None:
  print("Нет корней")
elif x1==x2:
  print(f"x1=x2={x2}")
else:
  print(f"x1={x1}, x2={x2}")
````

## Задача: Високосный год

Требуется определить, является ли год високосным. В соответствии с григорианским календарем, год является високосным, если его номер кратен 4, но не кратен 100, а также если он кратен 400.

````python
y = 2000

def divby(y,n):
  return y%n==0

def leapyear(y):
  return (divby(y,4) and not divby(y,100)) or divby(y,400)

if leapyear(y):
  print('Високосный')
else:
  print('Не високосный')
````

````python
import jturtle as t

def poly(x,n):
  for _ in range(n):
    t.forward(x)
    t.right(360/n)

def square(x):
  poly(x,4)

def triangle(x):
  poly(x,3)

def house(x):
  square(x)
  t.forward(x)
  t.right(30)
  triangle(x)
  t.left(30)
  t.right(180)
  t.penup()
  t.forward(x//3)
  t.left(90)
  t.forward(x//3)
  t.pendown()
  square(x//3)

for _ in range(4):
  house(100)
  t.penup()
  t.forward(100)
  t.right(90)
  t.forward(2*100//3)
  t.left(90)
  t.pendown()
  t.left(90)

t.done()
````

## Напечатать простые числа от 2 до 100


````python
def isprime(n):
  for i in range(2,n):
    if n%i==0:
      return False
  return True

for i in range(2,101):
  if isprime(i):
    print(i,end=' ')
````

````python
def factor(n):
  i = 2
  while n>1:
    if n%i==0:
      n = n//i
      print(i,end=' ')
    else:
      i+=1

factor(13)
````

````python
for i in range(2,100):
  print(f"{i} -> ",end='')
  factor(i)
  print()
````

## Число нулей в конце 100!

````python
def fact(n):
  f = 1
  for i in range(2,n+1):
    f = f*i
  return f

def count_zeros(n):
  r = 0
  while n%10==0:
    r+=1
    n//=10
  return r

count_zeros(fact(100))
````

````python
int(float(fact(100))/10)
````

````python
def sumdigits(x,base=10):
  s = 0
  while x>0:
    s+=x%base
    x//=base
  return s

sumdigits(13,2)
````

````python
def printd(x, base=10):
  k = 1
  while k*base<x:
    k*=base
  while k>0:
    print(x//k,end='')
    x=x%k
    k//=base
  print()

printd(13,2)
````
